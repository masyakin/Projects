```python
def display_df(df):
    """ Функция переименует столбцы в соответствии со змеиным регистром и покажет:
    - произвольные 5 строк датафрейма
    - типы данных
    - пропуски
    - дубликаты"""
    
    def to_snake_case(name):
        name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
        name = re.sub('__([A-Z])', r'_\1', name)
        name = re.sub('([a-z0-9])([A-Z])', r'\1_\2', name)
        return name.lower()

    df.columns = df.columns.to_series().apply(to_snake_case)
    print('')
    cprint('Первые 5 строк датафрейма', 'blue', attrs=['blink'])
    display(df.head())
    print('')
    print('')
    cprint('Типы данных', 'blue', attrs=['blink'])
    print('')
    print(df.info())
    print('')
    print('')
    cprint('Пропуски', 'blue', attrs=['blink'])
    print('')
    count = 0
    for element in df.columns:
        if df[element].isna().sum() > 0:
            print(element, '-', df[element].isna().sum(), 'пропусков')
            count =+1
    if count == 0:
        print('Пропусков нет')
    print('')
    print('')
    cprint('Дубликаты', 'blue', attrs=['blink'])
    print('')
    if df.duplicated().sum() > 0:
        print('Дубликатов:', df.duplicated().sum())
    else:
        print('Дубликатов нет')   
```

Посмотрим какие признаки выделит важными модель машинного обучения, затем с помощью инструмента shap проиллюстрируем как каждая фича влияет на отток. Для этого обучим модель CatBoost. 

```python
#разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = df.drop('churn', axis=1)
y = df['churn']
```

```python
# разделяем модель на обучающую и тестовую выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, 
                                                    random_state = 13, stratify=y)
```

```python
# разделяем модель на валидационную и тестовую выборку
X_val, X_test, y_val, y_test = train_test_split(X, y, test_size=0.5, 
                                                    random_state = 13, stratify=y)
```

Проверим соотношение отточных клиентов в выборках

```python
# Зададим параметры для модели
params = {'verbose':100,
         'random_seed':13,
         'learning_rate':0.9}
```

```python
# Задаем алгоритм для нашей модели

model = CatBoostClassifier(**params)
```

```python
df_2_product = df.query('products == 2').copy()
```

```python
# Сверим дисперсии выборок

cprint("Дисперсия среднего дохода для 1 продукта равна {:.0f}"
       .format(np.var(df_1_product['estimated_salary'], ddof=1)), 'grey', attrs=['blink'])
cprint("Дисперсия среднего дохода для 2 продуктов равна {:.0f}"
       .format(np.var(df_2_product['estimated_salary'], ddof=1)), 'grey', attrs=['blink'])
```

```python
cprint("Отношение дисперсий составляет {:.0%}"
       .format((np.var(df_1_product['estimated_salary'], ddof=1)) / 
               (np.var(df_2_product['estimated_salary'], ddof=1))), 'grey', attrs=['blink'])
```

Дисперсии выборок не сильно различаются, значит будем использовать параметр `equal_var = True`

```python
sample_1 = df_1_product['estimated_salary']
sample_2 = df_2_product['estimated_salary']

# критический уровень статистической значимости
# если p-value окажется меньше него - отвегнем гипотезу
# применим поправку Бонферрони и поделим уровень значимости на 2 - количество тестов.
alpha = .01/2

results = st.ttest_ind(sample_1, sample_2, equal_var = True)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

`p-значение` равно 0.41, что не дает основания отвергнуть предположение об отсутствии различий между доходами клиентов, пользующихся 1 и 2 продуктами, т.е. маловероятна случайность результатов теста гипотезы.

Таким образом, средний доход клиентов, пользующихся 1 продуктом, не отличается от среднего дохода клиентов, пользующихся 2 продуктами: разница статистически не значима (p-значение: 0,41) и составляет 965: 99 487 - у клиентов, пользующихся 1 продуктом банка против 100 452 -   у клиентов, пользующихся 2 продуктами банка.

### Уровень оттока среди клиентов старше 55 лет, которые пользуются двумя продуктами банка, равен уровню оттока среди клиентов, которые пользуются одним продуктом

Сформулируем нулевую гипотезу и альтернативную гипотезы:

H0: "средний уровень оттока среди клиентов старше 55 лет, которые пользуются одним и двумя продуктами банка равны".

Н1: "средний уровень оттока среди клиентов старше 55 лет, которые пользуются одним и двумя продуктами банка не равны".

```python
df_1_product = df.query('products == 1 and age > 55').copy()
```

```python
df_2_product = df.query('products == 2 and age > 55').copy()
```

```python
# Сверим дисперсии выборок

cprint("Дисперсия среднего оттока для 1 продукта равна {:.2f}"
       .format(np.var(df_1_product['churn'], ddof=1)), 'grey', attrs=['blink'])
cprint("Дисперсия среднего оттока для 2 продуктов равна {:.2f}"
       .format(np.var(df_2_product['churn'], ddof=1)), 'grey', attrs=['blink'])
```

```python
cprint("Отношение дисперсий составляет {:.0%}"
       .format((np.var(df_1_product['score'], ddof=1)) / 
               (np.var(df_2_product['score'], ddof=1))), 'grey', attrs=['blink'])
```

Дисперсии выборок сильно различаются, значит будем использовать параметр `equal_var = False`

```python
sample_1 = df_1_product['churn']
sample_2 = df_2_product['churn']

# критический уровень статистической значимости
# если p-value окажется меньше него - отвегнем гипотезу
# применим поправку Бонферрони и поделим уровень значимости на 2 - количество тестов.
alpha = .01/2

results = st.ttest_ind(sample_1, sample_2, equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

`p-значение` меньше критического уровня значимости, что дает основание отвергнуть предположение об отсутствии различий между уровнем оттока среди клиентов старше 55 лет, пользующихся 1 и 2 продуктами, т.е. маловероятна случайность результатов теста гипотезы.

Таким образом, статистический анализ подтверждает, что средний уровень оттока среди клиентов старше 55 лет, пользующихся 1 продуктом, выше среднего среднего уровня оттока среди клиентов, пользующихся 2 продуктами: разница статистически значима и составляет 23 процентных пункта: 43% - среди клиентов, пользующихся 1 продуктом банка против 20% - среди клиентов, пользующихся 2 продуктами банка.
